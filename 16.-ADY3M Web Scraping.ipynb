<h2 id="DSCW">Downloading da Scraping data daga Web Page (shafi)</h2> 


Farko zamuyi dowloading abinda ke cikin web page <br>
zamu fara saka url dinmu a variable <br>
sannan muyi amfani da library din requests:


````python
url = "https://ady3m.github.io/ady3mtestwebsite.github.io/"
````

se muyi amfani da kalmar  <code>get</code> muyi downloading data din cikin web page din a text format. se mu ajiye a variable me suna <code>data</code>:


````python
data  = requests.get(url).text
````

se mu hada <code>BeautifulSoup</code> object tare d amfani da <code>BeautifulSoup</code> constructor


````python
soup = BeautifulSoup(data,"html5lib")  # zamu hada soup object me suna soup da variable din 'data'
````

Muyi scraping links na websites


````python
for link in soup.find_all('a',href=True):  # a html anchor/link ana saka masa tag <a>

    print(link.get('href'))
````

Muyi scraping duka tags masu hotuna


````python
# mu duba links masuhotuna
for link in soup.find_all('img'):# a html ana sakawa hotuna tag din <img>
    print(link)
    print(link.get('src'))
````

### Mu ajiye links din hotunan da muka samu a cikin list

````python
# mu fara hada empty list da zamu ajiye links din hotunan
links_hotuna =[]
for link in soup.find_all('img'):# a html ana sakawa hotuna tag din <img>

#zamu iya ajiye su a cikin list dinmu
    src = link.get('src')
    links_hotuna.append(url+src.strip())
````

````python
# mu duba links din hotuna da mukayi scraping
links_hotuna
````

### Mu hada folder da zamu saka hotunan

````python

import os
directory_path = "C:/Users/hp/Downloads/download/scraped hotuna"

# Mu duba ko da akwai shi kafin mu gina wani
if not os.path.exists(directory_path):
    # If it doesn't exist, create the directory
    os.makedirs(directory_path)
````

````python
import os
    # Mu dakko hoton
for i in links_hotuna:
    image_response = requests.get(i)
        # Mu duba ko connection din yayi
    if image_response.status_code == 200:
                # Mu dakko sunan file din daga URL
        filename = os.path.join(directory_path, os.path.basename(i))
        # Muyi saving hoton a folder din da muka hada
        with open(filename, 'wb') as image_file:
            image_file.write(image_response.content)
            print(f"Munyi Downloading: {filename.strip('C:/Users/hp/Downloads/download/scraped hotuna')}.")
    else:
        print(f"Failed to download image from {i}")
````
